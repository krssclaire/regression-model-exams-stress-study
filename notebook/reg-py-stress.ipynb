```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# load data
survey_data_path = '../dataset/dati-sondaggio-stress.csv'
survey_data = pd.read_csv(survey_data_path)

# define parameters X, y
y = survey_data.LivelloStress
X = survey_data[['EsamiSem', 'Genere']]

# define model
model = LinearRegression()

# fit model
model.fit(X, y)

# grid (piano della regressione)
gender_range = np.linspace(X['Genere'].min(), X['Genere'].max(), 10)
esami_sem_range = np.linspace(X['EsamiSem'].min(), X['EsamiSem'].max(), 10)
f_grid, q_grid = np.meshgrid(gender_range, esami_sem_range)
x_grid = np.c_[f_grid.ravel(), q_grid.ravel()]
y_pred_grid = model.predict(x_grid).reshape(f_grid.shape)

# setup 3d graph
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# real points
ax.scatter(X['Genere'], X['EsamiSem'], y, color='orange', label='Osservazioni')

# Piano di regressione
ax.plot_surface(f_grid, q_grid, y_pred_grid, color='sandybrown', alpha=0.5)

# ticks
ax.set_xlabel('Genere (D)')
ax.set_ylabel('Esami al semestre (X)')
ax.set_zlabel('Livello di stress (Y)')

plt.legend()
plt.show()
```
